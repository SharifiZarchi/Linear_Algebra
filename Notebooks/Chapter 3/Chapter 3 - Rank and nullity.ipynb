<div class="cell rtl-align" dir="rtl">

\begin{equation}
    A = \begin{bmatrix}1 & 1 & 2 \\ 2 & 1 & 3 \\  3 & 1 & 4 \\ 4 & 1 & 5\end{bmatrix}
\end{equation}

حال می‌خواهیم بدانیم که چه هنگامی معادله‌ی $Ax=b$ جواب خواهد داشت. از‌آنجایی که ماتریس $A$ یک ماتریس $4\times 3$ است، ما ۴ معادله و ۳ مجهول خواهیم داشت. به همین علت امکان دارد که معادله‌ی داده شده جواب نداشته باشد. اما آیا این همیشه همینگونه است؟ به مثال زیر توجه کنید، آیا می‌توانید پاسخی برای این معادله پیدا کنید؟

\begin{equation}
    Ax = \begin{bmatrix}1 & 1 & 2 \\ 2 & 1 & 3 \\  3 & 1 & 4 \\ 4 & 1 & 5\end{bmatrix} \begin{bmatrix}x_1 \\ x_2 \\ x_3 \end{bmatrix} = \begin{bmatrix} 1 \\ 2 \\ 3 \\4 \end{bmatrix} = b
\end{equation}

<details>
<summary> <font color='green'>برای مشاهده‌ی پاسخ کلیک کنید </font></summary>
می‌توان رابطه‌ی داده شده را به صورت زیر بازنویسی کرد:

\begin{equation}
    x_1 \begin{bmatrix}1 \\ 2 \\ 3 \\ 4 \end{bmatrix} + x_2 \begin{bmatrix} 1 \\ 1 \\ 1 \\ 1 \end{bmatrix} + x_3 \begin{bmatrix} 2 \\ 3 \\ 4 \\ 5 \end{bmatrix} = \begin{bmatrix} 1 \\ 2 \\ 3 \\4 \end{bmatrix}
\end{equation}

بدیهیست که یک پاسخ قابل قبول برای این مسئله $x = \begin{bmatrix} 1 \\ 0 \\ 0 \end{bmatrix}$ است. با کمی دقت متوجه می‌شویم که یک پاسخ قابل قبول دیگر $x = \begin{bmatrix} 0 \\ -1 \\ 1 \end{bmatrix}$ می‌باشد.
به عبارتی این معادله تنها هنگامی دارای پاسخ است که $b \in C(A)$ باشد، یعنی بردار پاسخ درون فضای برداری ماتریس ضرایب قرار بگیرید.

دستگاه معادلات همگن زیر را در نظر بگیرید:

\begin{equation}
    x_1 -3x_2 - 2x_3 = 0 \\
    -5x_1 + 9x_2 + x_3 = 0
\end{equation}

دستگاه معادلات فوق را می‌توانیم به فرم ماتریسی $Ax = 0$ بنویسیم که در آن:

\begin{equation}
  A = \begin{bmatrix} 1 & -3 & -2 \\ -5 & 9 & 1 \end{bmatrix}, \quad
  x = \begin{bmatrix} x_1 \\ x_2 \\ x_3 \end{bmatrix}
\end{equation}

</div>

<link rel="stylesheet" href="style.css">

## تعریف فضای پوچی
فضای پوچی ماتریس $A_{m\times n}$، که آن را به صورت Nul A می‌نویسیم، مجموعه‌ی تمام جواب‌های معادله‌ی همگن $Ax=0$ است که آن را به زبان ریاضی به صورت زیر تعریف می‌کنیم:

\begin{equation}
    \text{Nul} ~ A = \{ x: x \text{ is in } \mathbb{R}^n \text{ and } Ax = 0\}
\end{equation}

</div>

<link rel="stylesheet" href="style.css">
<div class="cell rtl-align" dir="rtl">

### پرسش
بررسی کنید که آیا بردار $u = \begin{bmatrix} 5 \\ 3 \\ -2 \end{bmatrix}$ به فضای پوچی ماتریس $A$ که آن را پیش‌تر معرفی کردیم تعلق دارد یا نه.


<details>
<summary> <font color='green'>برای مشاهده‌ی پاسخ کلیک کنید </font></summary>

برای بررسی اینکه آیا $u$ شرط $Au=0$ را برقرار می‌کند کافیست که محاسبات زیر را انجام دهیم:

\begin{equation}
    Au = \begin{bmatrix} 1 & -3 & -2 \\ -5 & 9 & 1 \end{bmatrix}
    \begin{bmatrix} 5 \\ 3 \\ -2 \end{bmatrix} =
    \begin{bmatrix} 5 - 9 + 4 \\ -25 + 27 -2 \end{bmatrix} =
    \begin{bmatrix} 0 \\ 0 \end{bmatrix}
\end{equation}

پس $u$ به فضای پوچی این ماتریس تعلق دارد.

# بدست آوردن فضای پوچی با rref


</div>

```python
# @title Row Reduced Echelon Form
# https://gist.github.com/sgsfak/77a1c08ac8a9b0af77393b24e44c9547
def rref(A, tol=1e-8, debug=False):
  R = A.copy().astype('float64')
  rows, cols = R.shape
  r = 0
  pivots_pos = []
  row_exchanges = np.arange(rows)
  for c in range(cols):
    if debug: print("Now at row", r, "and col", c, "with matrix:\n", R)

    ## Find the pivot row:
    pivot = r
    for i in range(r, rows):
      if R[i, c] != 0:
        pivot = i
        break

    m = np.abs(R[pivot, c])
    if debug: print("Found pivot", m, "in row", pivot)
    if m <= tol:
      ## Skip column c, making sure the approximately zero terms are
      ## actually zero.
      R[r:rows, c] = np.zeros(rows-r)
      if debug: print("All elements at and below (", r, ",", c, ") are zero.. moving on..")
    else:
      ## keep track of bound variables
      pivots_pos.append((r,c))

      if pivot != r:
        ## Swap current row and pivot row
        R[[pivot, r], c:cols] = R[[r, pivot], c:cols]
        row_exchanges[[pivot,r]] = row_exchanges[[r,pivot]]

        if debug: print("Swap row", r, "with row", pivot, "Now:\n", R)

      ## Normalize pivot row
      R[r] /= R[r, c]
      ## Eliminate the current column
      v = R[r, c:cols]
      ## Above (before row r):
      if r > 0:
        ridx_above = np.arange(r)
        R[ridx_above, c:cols] = R[ridx_above, c:cols] - np.outer(v, R[ridx_above, c]).T
        if debug: print("Elimination above performed:\n", R)
      ## Below (after row r):
      if r < rows-1:
        ridx_below = np.arange(r+1,rows)
        R[ridx_below, c:cols] = R[ridx_below, c:cols] - np.outer(v, R[ridx_below, c]).T
        if debug: print("Elimination below performed\n:", R)
      r += 1
    ## Check if done
    if r == rows:
      break;
  return R, pivots_pos, row_exchanges
```

```python
# @title permutation matrices
def row_perm(rows, row_exchanges):
  p = np.zeros((rows, rows))
  for idx in range(rows):
    p[idx][row_exchanges[idx]] = 1
  return p.T

def col_perm(cols, col_exchanges):
  p = np.zeros((cols, cols))
  for idx in range(cols):
    p[col_exchanges[idx]][idx] = 1
  return p.T
```

```python
# given the rref find the basis for nullspace
def find_nullspace(R, pivots_pos, debug=False):
  # we must first rewrite R as [[I, F], [0, 0]]
  # then the null space will be [[-F], [I]]
  rows, cols = R.shape
  rank = len(pivots_pos)
  # find the correct ordering of the columns
  col_exchanges = np.arange(cols)
  for i in range(rank):
    pivot_col = pivots_pos[i][1]
    col_exchanges[[pivot_col, i]] = col_exchanges[[i, pivot_col]]
    R[:, [pivot_col, i]] = R[:, [i, pivot_col]]
  if debug: print('reduced echelon form in [[I, F], [0, 0]] is:\n', R)
  F = R[:rank, rank:cols]
  N = np.concatenate([-F, np.identity(cols - rank)])
  if debug: print('Nullspace in [[-F], [I]] is:\n', N)
  # reorder rows of nullspace
  return col_perm(cols, col_exchanges) @ N
```

```python
A = np.array([[1, 2, 3], [2, 4, 6], [2, 6, 8], [2, 8, 10]])
R, pivots_pos, row_exchanges = rref(A.T)
find_nullspace(R, pivots_pos)
```

```python
A = np.array([[1, 0, 1, 3], [2, 3, 4, 7], [-1, -3, -3, -4]])
R, pivots_pos, row_exchanges = rref(A)
print(R, pivots_pos)
find_nullspace(R, pivots_pos)
```

```python
M = Matrix([[1, 0, 1, 3], [2, 3, 4, 7], [-1, -3, -3, -4]])
print("Matrix : {} ".format(M))

# Use sympy.rref() method
M_rref = M.rref()

print("The Reduced Row Echelon Form of matrix M and the pivot columns : {}".format(M_rref))
```

#حل کامل معادلات خطی


</div>

<link rel="stylesheet" href="style.css">
<div class="cell rtl-align" dir="rtl">

اکنون با توجه به چیزی که در ۲ بخش پیش آموختیم تلاش می‌کنیم تا معادلات به فرم $Ax=b$ را به طور کامل حل کنیم. پاسخ کلی این فرم از معادلات را می‌توان به صورت زیر نوشت:

\begin{equation}
    Ax = b  \rightarrow \begin{cases}\text{particular solution} & Ax_p = b \\ \text{homogenous solution} & Ax_n=0\end{cases} \rightarrow A(x_p +x_n) = b
\end{equation}

که در این معادله $x_p$ یک جواب خاص این معادله و $x_n$ متعلق به فضای پوچی ماتریس است.

 معادله‌ی زیر را در نظر بگیرید:

\begin{equation}
    Ax = \begin{bmatrix}1 & 2 & 2 & 2\\ 2 & 4 & 6 & 8 \\  3 & 6 & 8 & 10 \end{bmatrix} \begin{bmatrix}x_1 \\ x_2 \\ x_3 \\ x_4 \end{bmatrix} = \begin{bmatrix}b_1 \\ b_2 \\ b_3\end{bmatrix}
\end{equation}

حال با تشکیل ماترسی افزایش یافته معادله‌ی فوق آن را حل می‌کنیم:

\begin{equation}
    \text{Augmented Matrix} = \begin{bmatrix} A & b \end{bmatrix} =
    \left[ \begin{matrix} 1 & 2 & 2 & 2 \\ 2 & 4 & 6 & 8 \\ 3 & 6 & 8 & 10 \\
    \end{matrix}
    \left| \, \begin{matrix} b_1 \\ b_2 \\ b_3 \\
    \end{matrix} \right. \right]
\end{equation}

تلاش کنید تا به کمک حذف گاوسی ماتریس افزوده را ساده کنید. آیا می‌توانید شرط پاسخ داشتن این دستگاه معادلات را پیدا کنید؟

<details>
<summary> <font color='green'>برای مشاهده‌ی پاسخ کلیک کنید </font></summary>
داریم:

\begin{equation}
    \left[ \begin{matrix} 1 & 2 & 2 & 2 \\ 2 & 4 & 6 & 8 \\ 3 & 6 & 8 & 10 \\
    \end{matrix}
    \left| \, \begin{matrix} b_1 \\ b_2 \\ b_3 \\
    \end{matrix} \right. \right] \rightarrow \left[ \begin{matrix} 1 & 2 & 2 & 2 \\ 0 & 0 & 2 & 4 \\ 0 & 0 & 2 & 4 \\
    \end{matrix}
    \left| \, \begin{matrix} b_1 \\ b_2 - 2b_1 \\ b_3 -3b_1 \\
    \end{matrix} \right. \right] \rightarrow \left[ \begin{matrix} 1 & 2 & 2 & 2 \\ 0 & 0 & 2 & 4 \\ 0 & 0 & 0 & 0 \\
    \end{matrix}
    \left| \, \begin{matrix} b_1 \\ b_2 - 2b_1 \\ b_3 - b_1 - b_2 \\
    \end{matrix} \right. \right]
\end{equation}

از آخرین معادله نتیجه می‌گیریم که پاسخ داشتن یا نداشتن این دستگاه معادلات به برقرار شرط $b_3 - b_1 - b_2 = 0$ بستگی دارد.


</div>

<link rel="stylesheet" href="style.css">
<div class="cell rtl-align" dir="rtl">

تلاش کنید به ازای $b$ داده شده فرم کلی پاسخ این دستگاه معادلات را پیدا کنید.

\begin{equation}
    Ax = \begin{bmatrix}1 & 2 & 2 & 2\\ 2 & 4 & 6 & 8 \\  3 & 6 & 8 & 10 \end{bmatrix} \begin{bmatrix}x_1 \\ x_2 \\ x_3 \\ x_4 \end{bmatrix} = \begin{bmatrix}1 \\ 5 \\ 6\end{bmatrix}
\end{equation}
برای انجام اینکار می‌توانید متغیرهای آزاد را برابر با صفر قرار دهید و یک جواب خاص یا همان $x_p$ را بیابید.
<details>
<summary> <font color='green'>برای مشاهده‌ی پاسخ کلیک کنید </font></summary>
شرط $b_3 = b_1 + b_2 $ برقرار بوده و داریم:

\begin{equation}
    \left[ \begin{matrix} 1 & 2 & 2 & 2 \\ 0 & 0 & 2 & 4 \\ 0 & 0 & 0 & 0 \\
    \end{matrix}
    \left| \, \begin{matrix} 1 \\ 3 \\ 0 \\
    \end{matrix} \right. \right]  \xrightarrow{x_2=x_4=0} \begin{cases} x_1 + 2x_3 = 1\\ 2x_3 = 3 \end{cases} \rightarrow \begin{cases} x_1 = -2 \\ x_3 = \frac{3}{2} \end{cases} \rightarrow x_p = \begin{bmatrix} -2 \\ 0 \\ \frac{3}{2} \\ 0 \end{bmatrix}
\end{equation}

از آخرین معادله نتیجه می‌گیریم که پاسخ داشتن یا نداشتن این دستگاه معادلات به برقرار شرط $b_3 - b_1 - b_2 = 0$ بستگی دارد.


</div>

<link rel="stylesheet" href="style.css">
<div class="cell rtl-align" dir="rtl">

حال تلاش می‌کنیم برای یافتن تمامی پاسخ‌های قابل قبول معادله‌ی فوق، فضای پوچی را بدست آوریم و $x_n$ را پیدا کنیم.

\begin{equation}
    Ax = \begin{bmatrix}1 & 2 & 2 & 2\\ 2 & 4 & 6 & 8 \\  3 & 6 & 8 & 10 \end{bmatrix} \begin{bmatrix}x_1 \\ x_2 \\ x_3 \\ x_4 \end{bmatrix} = 0
\end{equation}
برای انجام اینکار می‌توانید متغیرهای آزاد را برابر با صفر قرار دهید و یک جواب خاص یا همان $x_p$ را بیابید.
<details>
<summary> <font color='green'>برای مشاهده‌ی پاسخ کلیک کنید </font></summary>
فرم پلکانی این معادله را که پیشتر حساب کردیم به یاد می‌آوریم:

\begin{equation}
    u=\begin{bmatrix} 1 & 2 & 2 & 2 \\ 0 & 0 & 2 & 4 \\ 0 & 0 & 0 & 0 \\
    \end{bmatrix}\rightarrow \begin{cases} x_2 = 1,~x_4 = 0 → & x_1 = -2 , ~ x_3 = 0\\ x_2 = 0,~ x_4 = 1 → & x_1=2, ~ x_3 =-2\end{cases}
\end{equation}

اکنون توانستیم که پایه‌های فضای پوچی را بدست آوریم. پس:

\begin{equation}
    x_n = c \begin{bmatrix}-2 \\ 1 \\ 0 \\ 0 \end{bmatrix} + d\begin{bmatrix}2 \\ 0 \\ -2 \\ 1 \end{bmatrix},\quad c,d \in \mathbb{R}
\end{equation}


</div>

```python
# @title solve Ax = b
def solve(A, b, debug=False):
  m, n = A.shape
  # x_p holds the particular solution
  x_p = np.zeros(n)
  # first we create the augmented matrix [A, b]
  augmented_matrix = np.concatenate([A, b], axis=1)
  if debug: print('Augmented Matrix [A b] is :\n', augmented_matrix)
  # find the reduced echelon form of the augmented matrix
  u, pivots_pos, row_exchanges = rref(augmented_matrix, debug=debug)
  # rank is equal to the number of pivots
  r = len(pivots_pos)
  # if the last column (b) has a pivot there are no solutions!
  if pivots_pos[-1][1] == n:
    print('There are no solutions!')
    return
  # non-pivot variables are free so we set them to zero
  pivot_cols = [c for _,c in pivots_pos]
  i = 0
  for r in pivot_cols:
    x_p[r] = u[i][-1]
    i += 1
  print(x_p)
  nullspace = find_nullspace(u[:, :-1], pivots_pos, debug=debug)
  print('m')
  print(nullspace)
  x_n = [column for column in nullspace.T]
  print('solution is:', x_p.T, '+', x_n)
```

```python
A = np.array([[1, 2, 2, 2], [2, 4, 6, 8], [3, 6, 8, 10]])
b = np.array([[1], [5], [6]])
solve(A, b)
```
